```python
import zipfile
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from io import StringIO
```

```python
z = zipfile.ZipFile('time-series.zip')

df = []
for i in z.infolist():
    df.append(pd.read_csv(StringIO(z.read(i.filename).decode('utf_8'))))
```

```python
for i in range(len(df)):
    df[i]['value'].plot()
    plt.show()
```
